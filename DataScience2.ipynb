```python
# Ensure 'Date' is in datetime format
dataset['Dated'] = pd.to_datetime(dataset["Dated"])

# Define time periods
pre_covid = dataset[dataset['Dated'] < '2020-03-01']
during_covid = dataset[(dataset['Dated'] >= '2020-03-01') & (dataset['Dated'] <= '2021-03-01')]

# Compare averages
print("Pre-Covid Unemployment Rate:", pre_covid['Estimated_Unemployed_Rate_India'].mean())
print("During-Covid Unemployment Rate:", during_covid['Estimated_Labour_particiption_Rate'].mean())
```

```python
import seaborn as sns

dataset['Period'] = dataset['Dated'].apply(lambda d: 'Pre-Covid' if d < pd.to_datetime('2020-03-01') else 'During-Covid')
sns.boxplot(data=dataset, x='Period', y='Estimated_Unemployed_Rate_India')
plt.title("Unemployment Rate Before vs During Covid")
plt.show()
```
